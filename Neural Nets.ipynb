```python
from sklearn.neural_network import MLPClassifier

df = pd.read_csv('PBP - 2016 - Week 1.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['homeTeam', 'awayTeam', 'homeId', 'offenseId', 'awayId', 'defenseId', 'offenseTeam', 'defenseTeam', 'year','description', 'wallclock'])
```

```python
def getWinner(gameId, df):
    final = df.query('gameId == @gameId').iloc[[-1]]
    if(int(final.homeScore) > int(final.awayScore)):
        return 'home'
    else:
        return 'away'
```

```python
# add a winner label
for index,row in df.iterrows():
    df.at[index,'winner'] = getWinner(row['gameId'], df)
```

```python
df.head()
```

```python
#need to get just the label
labels = []
for index,row in df.iterrows():
    labels.append(getWinner(row['gameId'], df))
```

```python
 clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)
```

```python
testing_copy = df.copy()
testing_copy = pd.get_dummies(testing_copy)
testing_copy = testing_copy.drop(columns=['winner_away', 'winner_home'])
```

```python

clf.fit(testing_copy, labels)
```

```python
for index,row in testing_copy.iterrows():

    pred = clf.predict(row.values.reshape(1,-1))
    real = getWinner(row['gameId'])
    print(pred[0] + ' vs ' + real)
    
```
